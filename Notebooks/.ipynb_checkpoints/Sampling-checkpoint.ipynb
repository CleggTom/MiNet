```julia
T=Int(1e2) #Gibbs steps
N=3 #n-1
s=125 #imposed sum

dX = Poisson()
dY = Poisson()
dR = Poisson()

function sample_BP(X,Y,r)
    X = rand(Poisson(X - r))
    Y = rand(Poisson(Y - r))
    R = rand(Poisson(r))

    return([X+R,Y+R])
end

function pdf_BP(x,y,X,Y,r)
   q = 0
    for k = 0:min(x,y)
        q += exp(-r-lX-lY) * (r^k * lX^(x-k) * lY^(y-k)) / (factorial(k) * factorial(x-k) * factorial(y-k))
    end
    return(q)
end

x = zeros(T,N,2)
[x[1,n,:] .= sample_BP(1.0,1.0,0.1) for n = 1:N]

for t = 2:T
    x[t,:,:] .= x[t-1,:,:]

    #loop over n samples
    for i = 1:N
        #get sample not including i
        ind = collect(1:n)
        filter!(x -> x != i, ind)
        
        #proposed update
        prop = x[t,i,:] .+ rand.([max(-x[t,i,1], x[t,i,1]-10):(x[t,i,1]+10),
                                  max(-x[t,i,2], x[t,i,2]-10):(x[t,i,2])+10])
#     #calculate acceptance prob
#     v = pdf(dP, prop) * pdf(dP, s - sum(x[t,ind]) - prop) / (pdf(dP,x[t-1, i]) * pdf(dP, s - sum(x[t,])))
#     if rand() < v
#         x[t,i] = prop
#     end
    end
end
```
